## Detect outliers using IQR

25% 50% 75% 100%

IQR = q3 - q1

upper_bound = q3 + k + IQR

lower_bound = q1 - k * IQR

acceptable list of values for k: 1.5, 2, 3

```python
q1, q3 = df['spending_score'].quantile([0.25, 0.75])
```

```python
q1
```

```python
q3
```

```python
ss_iqr = q3 - q1
```

```python
ss_iqr
```

```python
k = 1.5

ss_upper = q3 + k + ss_iqr
ss_upper
```
